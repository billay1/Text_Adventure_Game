<h1>PUSS IN BOOTS</h1>


Are you a master of tricks? This game is for you! Welcome to the game "Puss in boots". You are a cat that were given as inheritance to the third son of a poor miller, while the first son inherited of the mill and the second son of a donkey. As your master complains on his inheritage, you reasure him that you will make him a rich man. You are in fact, a special cat: you can talk and you are clever! Clothed with a hat and a pair of red boots, you use of multiple tricks to reach your goal, making your master rich.

```python
"""
Created on Sun Oct 24 10:36 2021

@author: Billa.Yeyi
"""

"""
    DocString:
    
    A) Introduction:
    This game is based on the franco-italian story "The Puss in boots" written 
    in 1988 by the french Charles Perrault.  You are basically a cat, and your 
    mission is to help your master becoming rich. The game consists of four 
    stages and  has defined functions for start, stages and fail. 
    You will need some general knowledge of the story and answer correctly 
    questions to win as it is a game of tricks.

    
    Good luck!
    
    Huntint_Stage: Choose which animal you will hunt and offer the king
    River_stage: Choose which object you will use to block the king's carriage
    Guardian_stage: Use your cleverness to answer the guardian's questions and pass the gate 
    Ogre_stage: Use your cleverness to trick the ogre and win his castle
    
    B) Known Bugs and/or Errors:
    None.
    
"""
#Import packages
import random 
import time

#defining functions fail and exit

def fail():
    #defining fail function purpose
    """This function ends the game for any wrong move of the player. However, 
    the player has the option to play again"""
    
    print("GAME OVER")
    input(prompt="<Press Enter to Continue>\n")
    
    while True:
        decision= input(prompt= "Would you like to play again? Type yes or no.\n")
        try:
            #player will be able to play again
            if 'yes' in decision.lower():
                start()
                river_stage()
                guardian_stage()
                ogre_stage()
                break

            elif 'no' in decision.lower():
                #player will end and exit the game
                print("""
                Thank you for playing. See you soon!
                """)
                break
            else:
                print("Invalid command. Please try again.")
                input(prompt="<Press enter to continue>\n")
                continue #player will enter choice again because of incorrect input
                
        except:
                print("Invalid command. Please try again.")
                input(prompt="<Press enter to continue>\n")
                continue #player will enter choice again because of incorrect input

def exit():
    #defining the function exit purpose
    """This function ends the game after the player won. It also gives the player 
    the option to play again"""
    
    decision= input(prompt= "Would you like to play again? Type yes or no.\n")
    
    if 'yes' in decision.lower():
        #player will be able to play again
        start()
        river_stage()
        guardian_stage()
        ogre_stage()
        
    elif 'no' in decision.lower():
        #player will end and exit the game
        print("Thank you for playing. See you soon!")
        
    else:
        print("Invalid command. Please try again.")
        input(prompt="<Press enter to continue>\n")
        exit() #user will be redirected to exit because of wrong entry
        
        
# I)The Hunting_stage
def start() :
    #defining the start function purpose
    """This function introduces the player to game and the first stage."""
    
    print("""Story_stage_1:'The hunting stage'\n""") #first stage of the game
    
    #welcome message and guidlines
    print("""
    Welcome to the game "Puss in Boots"! As your master passed you some clothes
    and a pair of red boots, you decided to go hunting to offer meat to the king 
    as he loves eating meat. You give the offering on behalf of your Master 
    that you named to the king "Marquess of Carabas". 
    """)
    
    #choice specifications
    print("Be careful! You need to hunt an animal that does not have red meat as the King suffer from gout.")

    input(prompt="<Press enter to continue>\n")
    #list of animals to choose from
    list_animals = ['ox', 'buffalo', 'goose', 'mouse']
    print(list_animals)
    
    #user input
    choice= input(prompt="Which animal from the list will you hunt to please the King?\n")
    #correct choice
    if 'goose' in choice.lower():
        
        print(f"""
        Great choice! The King said "{list_animals [2]} is my favorite meat!" 
        and praises your master for the generous act.
        """)
        
        #player will be direted to stage_2 of game upon correct answer
        input(prompt="<Press enter to continue>\n")
        time.sleep(2)

    #wrong choices
    elif 'ox' in choice.lower() or 'buffalo' in choice.lower() or 'mouse'in choice.lower():
        print("""
        Bad choice! :( 
        The king got sick. He killed you and offered you to its Fox and sent your 
        master to prison.
        """)
        fail() #player fails the game because of wrong answer
        
    #in case user inputs wrong entry 
    else:
        print("""
        Wrong entry! Try again to choose an animal from the list/
        pay attention to the wording.\n""")
        input(prompt="<Press enter to continue>\n")
        start()
        river_stage()
        guardian_stage()
        ogre_stage()
        

start() #actual start of start function

# II) River_stage
print("""Story_stage_2: 'The river stage'\n""")
time.sleep(1)
print('*'*80)
#Guidlines on the second step to the player
print("""
You heard that the King is coming to visit your master with his daughter.
However, your master, supposedly the "Marquess of Carabas" does not have 
appropriate clothes and a casttle to welcome the King and your lies may be discovered.
But, you heard the Kings'carriage will pass by a river. You then suggest 
to your master to remove all his clothes and pretend to drown. Your objectif 
is to stop the King's carriage so they can rescue your master.\n""")
print('*'*80)

#Defining function river_stage
def river_stage():
    #defining the river_stage purpose
    """This function introduces the player to the 2nd stage: 'the river' """
    
    #guidlines on user choice
    print("""
    You need to block the road with an object so the carriage stops. Choose 
    between the following elements, the one that will most likely stops the carriage.
    1. master's old clothes
    2. 2 inches rock
    3. yourself waving
    """)
    
    while True:
        #user input
        guess= input(prompt= "Which one do you think will stop the carriage? \n")
        
        #correct guess
        if 'yourself waving' in guess.casefold():
            print("""
            Correct guess! If the carriage sees you from far away it will most likely 
            slow down and as the King recognizes you, he will order to stop.
            \n""")
            break
            
        #wrong alternative
        elif  "master's old clothes" in guess.casefold():
            print("""
            Wrong guess! The carriage will just trample on your master's clothes and 
            he will go back home naked and freezing.  
            \n""")
            time.sleep(2)
            
            #guess outcomes
            print("""
            The king's carriage couldn't find your castle.
            Your lies are discovered.You and your masterget get arrested. 
            Your master is thrown in prison while you are killed and cooked for the King's fox.
            """)
            break
            fail() #player fails the game because of wrong answer
            
        #wrong alternative
        elif  '2 inches rock' in guess.casefold():
            print("""
            Wrong guess! Think about it, a 2 inches rock is similar to a lemon size. 
            It can easily be overpassed by the carriage.
            \n""")
            time.sleep(2)
            
            #guess outcomes
            print("""
            The king's carriage couldn't find your castle. Your lies are discovered.
            Your master is thrown in prison while you are killed and cooked for the King's fox.
            \n""")
            break
            fail() #player fails the game because of wrong answer
            
        #in case user inputs wrong entry
        else:
            print("Wrong entry! Pay attention to the wording.")
            input(prompt="<Press enter to continue>\n")
            continue #player will enter choice again because of incorrect input
            
river_stage()

# after successfully having stopped the King's carriage
time.sleep(2)
print("""
The king's carriage stopped and your master was rescued and given new clothes. 
You explain to 
the king that some thieves stole your master's carriage and clothes. 
In the meanwhile, your master encounter the king's daughter: it is love
at first sight... ;)
\n""")
#puss leaves to prepare next plan: obtaining a castle to welcome the king
print("""
You excuse yourself to go before so you can prepare a warm welcome to the 
King at the Marquess of Carabas'castle while your master goes with the king 
in his carriage. As your master starts to worry because he does not own a castle,
you wink at him to tell him that you have a plan.\n
""")
#User little guess, wrong guess won't impact the story's flow
input(prompt= "What do you think you will do?\v\n") #user can input anything

print("""
Impressive idea! In fact, you heard that an ogre owns a castle and can transform 
into an animal and indicated to the King the way to the ogre castle, pretending
it was your master's castle.
You decide then to go pay the orgre a visit.\n
""")
# III) Guardian_Stage
print('*'*80)
print("Story_stage_3: 'The guardian stage'\n")
print('*'*80)
#meeting with the guardian in front of the ogre's castle gate
print("You arrived at the orgre's castle and the gate is kept by a guardian.\n")
print("You will be given the opportunity to have a personalised discussion with the guardian.\n")
input(prompt="<Press 'Enter' to continue>\n")

#Conversation between the cat and the guardian
print("The Guardian:\nHey! Halt! Who are you?\v")
input(prompt= "You: \t") #user can input anything
print("The Guardian:\nWhat are you looking for?\v")
input(prompt= "You: \t") #user can input anything
print("""The Guardian:\nI heard of you, the cat who can talk. They say that 
You are clever. Well we will check that. If you answer correctly my
question, then I will let you enter the castle. Ready?\v""")

input(prompt="<Press 'Enter' to start>\n")

#The guardian Questions
def guardian_stage():
    #defining guardian_stage purpose
    """This function allows the guardian to ask the puss a question, which the later
    has to answer correctly to enter the castle. The question will be randomely selected"""
    
    print("""\nYou have to answer correctly to pass.\n""")


    #list of questions from which one will be randomely selected and asked by guardian
    
    question_list = ['What is the word that can be formed from the following letters:\nt-a-e-r-g\n?',
    "I twinkle.There are millions of me.I will light your way.I come out at night. I hide in the city. Who am I?\n",
    'In which continent is the country Kuwait located?',
    'Who was the  president in South-Africa who fought against the appartheid?',]
        
    questions = 1 #user will have to answer only one question

    while questions > 0:
        question = random.choice(question_list)
        input(prompt="<Press enter to continue>\n")

        if question == question_list[0]:
            questions -=1

        #guardian's question number 1
            print(f"""
            Guardian:\n{question_list[0]}\n?
            """)
            chances = 2 #player will have 2 chances to answer in case answer is wrong
            answer = 'great'
            while True:
                guess = input(prompt="> \n")
                
                #right guess
                if guess.lower() == answer :
                    
                    print("""
                    Guardian: Excellent! Seems that the rumors are quite true. But let's 
                    see how you do for next question.
                    """)
                    print(f"""\nGood response!\n """)
                    break
                
            #Player will be directed to question_2 
            
            #wrong guess
                elif guess.lower() != answer:
                    chances -= 1

                    if chances > 0:
                        print(f"""
                        Guardian: Wrrroonng! Ahaha I got you. I will leave you {chances} other chance
                        to guess right.\n
                        """)
                        print(f"""\nBad response!\n""")
                        
            #Player will have the possibility to guess again, until there is no more chance
                    elif chances == 0:
                        print(f"""\n
                        Guardian: Double wrong! Got you again ahaha. Your chances are over. 
                        The correct guess was '{answer}'.Let see if you can answer 
                        next question correctly.
                        """)
                        print(f"""Bad response! \n""")

                        break
                        fail() #if no more chances to answer, the player fails the game
                    
                    #in case user inputs wrong entry
                    else:
                        print("Something went wrong. Please try again and pay attention to the wording")
                        input(prompt="<Press enter to continue>\n")
                        continue #player will enter choice again because of incorrect input
                        
                #in case user inputs wrong entry
                else:
                    print("Invalid command. Please try again.")
                    input(prompt="<Press enter to continue>\n")
                    continue #player will enter choice again because of incorrect input
        #question_2
        elif question == question_list[1]:
            questions -=1
            
            #guardian's question_2
            print(f"""\nGuardian:\n {question_list[1]}""")

            time.sleep(1)
            #answer choices
            print("""
            1. donald trump
            2. candle
            3. star
            4. torch
            """)
            print("Enter what you think is the good answer")
            choice = input(prompt=">  \n") #player choice entry

            #right guess
            if 'star' in choice.casefold():
                
                print(f"""Good response!\n""")
                print(f"""
                Guardian:\n I have to admit that you are indeed clever! 
                \n""") #player will be directed to stage_4
                break
            elif "donald trump" in choice.casefold() or "candle" in choice.casefold() or "torch" in choice.casefold():
                
                print(f"""Bad response! Correct answer was "star". 
                    \n
                    """)
                break
                fail() #player fails the game because of wrong answer
            
            #in case user inputs wrong entry
            else:
                print("Invalid command. Please try again.")
                input(prompt="<Press enter to continue>\n")
                continue #player will enter choice again because of incorrect input
                
        #question 3 in random choice
        elif question == question_list[2]:
            questions -=1
            
            print(f"""Guardian:\n {question_list[2]}?
                         1. europe
                         2. asia
                         3. africa
                """)
            choice = input(prompt=">  \n") #user entry
            
            #right guess
            if  "asia" in choice.casefold():
                print(f"""Good response!\n""")

                time.sleep(1)
                print(f"""
                Guardian:\n I have to admit that you are indeed clever!\n
                """)
                break
            #player will now enter the castle and meet the ogre
            #this will mark the start of stage_4
            
            #wrong guess
            elif 'europe' in choice.casefold() or "africa" in choice.casefold():
                print(f"""Bad response! Correct answer was "asia".\n""")

                time.sleep(1)
                print("""
                Guardian:\nWell, you are not as clever as I thought.
                I won't let you enter. Go and hunt some mouses ahaha.\n
                """)
                print("""You were not able to get into the castle. The carriage 
                arrived and the king found out your lies. The ogre, alerted by all the noise comes 
                out and eat you all. Everybody dies.
                """)
                break
                fail() #player fails the game after wrong answer
            
            #in case user inputs wrong entry
            else:
                print("Invalid command. Please try again.")
                input(prompt= "<Press enter to continue>\n")
                continue #player will enter choice again because of incorrect input
        
        
                
guardian_stage() #actual start of guardian stage 

#Guardian giving puss authorization to enter castle after correct guess
print(f"""
Guardian:\nyou can now enter. Good luck with not 
getting eaten by the ogre.\n
""")
# IV) Ogre_Stage
input(prompt="<Press enter to continue>\n")
print('*'*80)
print("Story_stage_3: 'The guardian stage'")
print('*'*80)
#puss enters the castle 
#meeting of the puss with the ogre
print("""\nCongratulation on getting into the castle! As you wander in, you fall into the ogre.
As he intended to catch and eat you, you striked a conversation with him,
telling him that he is powerfull and that no one can resist him\n""")
print("""The ogre has the ability to transform into any animal. You are aware of this 
ability and want to take advantage of it\n""")
time.sleep(1)
        
#Conversation with puss and ogre
print('*'*80)
print("Start of the conversation")
print('*'*80)
input(prompt="<Press 'enter' to continue>\n")
print("""Cat:\nOh my lord! You are the most powerfull of all! I have heard that 
you can transform into any animal and would love to see your power before 
dying in amazement of you
""") #puss talking to the ogre
time.sleep(1)
print("""Ogre:\nYou heard correct! I will grant your last wish before eating you.
In which animal do you want me to turn into?\v
""")#ogre response to puss
time.sleep(1)
#start of puss startegy to trick the ogre
print("""Cat:\n Oh my lord! I want to see you turn into a very large animal""")
time.sleep(1)
print("Ogre:\n As wish, as done\n")
time.sleep(1)
print('*'*80)
print("And the ogre transforms into an elephant")#ogre first transformation
print('*'*80)
time.sleep(1)
print("""Cat:\n Oh my lord! I am so frightened. Can you please turn into a smaller animal\v""")
time.sleep(1)
print("""Ogre:\n As wish, as done\n""")
time.sleep(1)
print('*'*80)
print("And the ogre transforms into a lion")#ogre second transformation
print('*'*80)
time.sleep(1)
print("""Cat:\n Oh my lord! I am so amazed but I am still frightened. 
Can you please turn into a smaller animal\v\n
""") #now start the tricks of the puss.
#player will have to do a smart move to defeat the ogre

#ogres_tage funtion definition
def ogre_stage ():
    #ogre_stage purpose
    """This function give a list of things you will want the ogre to transform into"""
    #player will have to choice in which animal should the ogre turn into
    print("""
    Choose in which animal you want the ogre to transform into. But be careful!
    As your choice may impact the course of the story. You have to make a clever choice!
    """)
    input(prompt="<Press enter to continue>\n")
    
    #list of possible choices of transformation
    list= ['orange', 'fox', 'mouse', 'cellphone'] 
    print(list)

    choice= input(prompt="Which one from the list do you choose\v \n") #player entry
    
    #right guess
    if 'mouse' in choice.lower():
        print("Correct choice!")
        time.sleep(1)
        print("Ogre:\n As wish, as done\n")
        time.sleep(1)
        #ogre fird transformation
        print(f"""
        The ogre transforms into a {list[2]}. As soon as he transforms, you 
        jump on him and devore him
        """)
        #right answer leads player to success, a happy ending and end of game
        
    #wrong guess
    elif 'fox' in choice.lower():
        print("Bad choice!")
        time.sleep(1)
        print("Ogre:\n As wish, as done\n")
        #ogre fird transformation
        time.sleep(1)
        print(f"""
        The ogre transforms into a {list[1]}. As soon as he transforms, he 
        jumps and devores you.The king's carriage arrives and the ogre devores 
        everybody including your master.
        """)
        fail() #player fails the game
        
    #wrong guess 
    elif 'orange' in choice.lower() or 'cellphone' in choice.lower():
        print("Bad choice!")
        time.sleep(1)
        #ogre can't transform because it is not an animal choice
        print("""Remember that the ogre can only transforms into an animal.
        As a result, the ogre thinks you are making fun of him. He jumps and 
        devores you.The king's carriage arrives and the ogre devores 
        everybody including your master. 
        """)
        fail() #player fails the game
    
    #in case user inputs wrong entry
    else:
        print("Invalid command. Please, pay attention to the wording and try again.")
        input(prompt="<Press enter to continue>\n")
        ogre_stage()#player will be directed to start of ogre_stage because of incorrect input
        
ogre_stage() #actual start of ogre_stage

#Defeat of the ogre
print("""
With the ogre defeat, you are now the new master of the castle and evry servant,
including the guardian acknowledge you as their new master. You tell them that their 
new master is named the 'Marquess or Carabas', your master, and that they should also obey him.
""")

#Arrival of the puss master's and the King. 

print("""
The king's carriage arrives and they are warmly welcome by the guardian and the servants.
The King is happy and after a sound meal, he decided to give his daughter 
in mariage to the Marquess of Carabas and give him some lands as property.\n
""")

#All stages completed 
#Happy ending for all
print("""You have successfully accomplished your mission of making your master rich!
He lives happily with his new wife and have lots of children as you decide 
to go explore the world for new exciting adventures...\n
""")

#End of the story
print('*'*80)
#congratulations message
print("""
Hurray!! You are the champion of tricks! Congratualtions on winning the game clever one!
Hope to see you soon
""")
print('*'*80)

print("Good bye ;)")

input(prompt="<Press enter to continue>\n")
exit() #user exit the game but have a possibility to start over
```
